```python
lista_tiendas = [tienda1, tienda2, tienda3, tienda4]
nombres_tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
```

#1. Análisis de facturación





```python
def A_Facturacion(datos, nombres):
  print('Análisis de facturación Por Tienda')
  facturacion_total = []
  for i in range(len(datos)):
    total = datos[i]['Precio'].sum()
    facturacion_total.append(total)
    print(f"{nombres[i]}: ${total:,.2f}")

  max_venta = max(facturacion_total)
  indice_mejor = facturacion_total.index(max_venta)
  print(f'Conclusión: La {nombres[indice_mejor]} es la que ha alcanzado la mayor facturación.\n')
  plt.figure(figsize=(10, 6))
  plt.bar(nombres, facturacion_total, color=['skyblue', 'salmon', 'lightgreen', 'orange'])
  plt.title('Comparativa de Facturación Total por Tienda')
  plt.ticklabel_format(style='plain', axis='y')
  plt.ylabel('Facturación ($)')
  plt.show()
```

```python
A_Facturacion(lista_tiendas, nombres_tiendas)
```

# 2. Ventas por categoría

```python
def V_Categoria(datos, nombres):
   print('Ventas por categoría')
   for i in range(len(datos)):
    populares = datos[i]['Categoría del Producto'].value_counts()
    top_cat = populares.idxmax()
    cant_cat = populares.max()
    print(f'{nombres[i]}: {top_cat} ({cant_cat} ventas)')
```

```python
V_Categoria(lista_tiendas, nombres_tiendas)
```

# 3. Calificación promedio de la tienda


```python
def C_prom(datos, nombres):
  print('Calificación promedio de la tienda')
  promedios_calif = []
  for i in range(len(datos)):
    promedio = datos[i]['Calificación'].mean()
    promedios_calif.append(promedio)

    status = ""
    if promedio >= 4.50:
      status = "Excelente"
    elif promedio > 4.00:
      status = "Buena"
    elif promedio >= 3.00:
      status = "Aceptable"
    else:
      status = "Mala"
    print(f'{nombres[i]}: {promedio:.2f} estrellas - ({status})')
```

```python
C_prom(lista_tiendas, nombres_tiendas)
```

# 4. Productos más y menos vendidos

```python
def vendidos(datos, nombres):
  print('Productos más y menos vendidos')
  for i in range(len(datos)):
      conteo_prod = datos[i]['Producto'].value_counts()
      mas_vendido = conteo_prod.idxmax()
      menos_vendido = conteo_prod.idxmin()
      print(f'{nombres[i]}:')
      print(f'  * Más vendido: {mas_vendido}')
      print(f'  * Menos vendido: {menos_vendido}')
```

```python
vendidos(lista_tiendas, nombres_tiendas)
```

# 5. Envío promedio por tienda

```python
def Envios(datos, nombres):
  print('Envío promedio por tienda')
  costos_envio = []
  for i in range(len(datos)):
    prom_envio = datos[i]['Costo de envío'].mean()
    costos_envio.append(prom_envio)
    print(f'{nombres[i]}: ${prom_envio:.2f}')
```

```python
Envios(lista_tiendas, nombres_tiendas)
```
